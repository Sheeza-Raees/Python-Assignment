```python
#Q5:
class car():
    def __init__(self,name,model,color,engine,body_type):
        self.name=name
        self.model=model
        self.color=color
        self.engine=engine
        self.body_type=body_type
##Method_1
    def updateColor(self, newColor):
        self.color= newColor
##Method_2
    def updateModel(self, newModel):
        self.model= newModel
##Method_3
    def updateEngine(self, newEngine):
        self.engine= newEngine
    
car1=car('Suzuki Swift',2012,'silver','1000cc','hatchback')
print("Car#1 Specifications are: \n")
print(car1.name)
print(car1.model)
print(car1.color, "- Old Version")
car1.updateColor("Pearl Arctic White")
print(car1.color,"- Modified Version")
print(car1.engine)
print(car1.body_type)
car2=car('Suzuki Cultus','2012','white','1000cc','hatchback')
print("\nCar#2 specifications are: \n")
print(car2.name)
print(car2.model, "- Old Version")
car2.updateModel(2016)
print(car2.model, "- Modified Version")
print(car2.color)
print(car2.engine)
print(car2.body_type)
car3=car('Mehran','2019','white','800cc','sedan')
print("\nCar#3 Specifications are: \n")
print(car3.name)
print(car3.model)
print(car3.color)
print(car3.engine,"- Old Version")
car3.updateEngine("1300cc")
print(car3.engine,"- Modified Version")
print(car3.body_type)
car4=car('Corolla','2007','Black','1000cc','hatchback')
print("\nCar#4 Specifications are: \n")
print(car4.name)
print(car4.model)
print(car4.color)
print(car4.engine)
print(car4.body_type)
car5=car('Mira', '2006', 'silver', '1000cc', 'hatchback')
print("\nCra#5 Specifications are: \n")
print(car5.name)
print(car5.model)
print(car5.color)
print(car5.engine)
print(car5.body_type)
```
